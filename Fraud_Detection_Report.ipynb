- Implementing a Balanced Oversampling Technique with a 1:1 Target Label Ratio<br><br>
To address the class imbalance in the dataset, I employed an oversampling technique to equalize the number of fraudulent (label 1) and non-fraudulent (label 0) transactions. After applying this method, the dataset was balanced with a total of ~600,000 records, comprising ~300,000 fraudulent and ~300,000 non-fraudulent transactions. This balance ensures that the machine learning model is trained on an evenly distributed dataset, improving its ability to accurately detect fraudulent activities. This method has lower recall score *0.65* for fradulent transactions, so it does not outperform SMOTE-based techniques.<br><br>
- Implementing an Oversampling Technique with a 10:1 Target Label Ratio<br><br>
I applied an oversampling technique to achieve a 10:1 ratio, resulting in a dataset with 1,000,000 non-fraudulent and 100,000 fraudulent transactions. By maintaining this ratio, I thought the model can better distinguish between fraudulent and non-fraudulent activities. However, it performed weaker than expected. While this method exhibits high accuracy *0.91* and recall *0.99* for non-fraudulent transactions, it fails to accurately identify fraudulent transactions with *0.144* recall score.<br><br>
- SMOTE with Logistic Regression<br><br>
Even though this model has respectable accuracy and f1 score, it demonstrates limitations in detecting fraudulent transactions, as evidenced by its moderate recall for Label 1.  While the precision for Label 1 is relatively high, the model's overall AUC-ROC score suggests there is room for improvement in distinguishing between fraudulent and non-fraudulent activities. Therefore, I have tried Decision Tree and Random Forest Models to improve it.<br><br>
- SMOTE with Decision Tree & Random Forest<br><br>
These models demonstrate strong overall performance, particularly in terms of AUC ROC, F1-score, and balanced recall and precision metrics for both labels.<br><br>
- Artificial Neural Network (ANN)<br><br>
Despite the model's high accuracy, it faces substantial challenges in identifying fraudulent transactions (Label 1), as reflected by its low recall and precision values for Label 1. Additionally, the model's significantly lower AUC-ROC score further underscores its inadequacy in distinguishing between fraudulent and non-fraudulent activities.


In Conclusion:

SMOTE-based models, particularly when combined with Decision Trees and Random Forests, exhibit superior performance in detecting credit card fraud, balancing precision and recall, and achieving high AUC ROC values. This highlights their effectiveness in handling class imbalances and improving detection accuracy.
